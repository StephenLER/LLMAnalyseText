```python
df = pd.read_pickle("/home/wangyuting/share/quant/wangyuting/liangjian/概念新闻数据/dataloader_keybert关键词_step3_2025_thres2.pkl")
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
mask = (
    (df['concept_code'] == '884126.WI') &
    (df['date'] >= '2025-01-01')
)

cols = ['date', 'title', 'content']
df_sub = df.loc[mask, cols].sort_values('date')

# 方案一：存成 CSV（方便人工查看）
df_sub.to_csv(
    "/home/wangyuting/share/quant/wangyuting/liangjian/llm4text/data/884126_2025_news.csv",
    index=False,
    encoding='utf-8-sig'   # Windows/Excel 友好一点
)

# 方案二（推荐给大模型）：存成 JSONL
df_sub.to_json(
    "/home/wangyuting/share/quant/wangyuting/liangjian/llm4text/data/884126_2025_news.jsonl",
    orient="records",
    lines=True,            # 每行一个 JSON
    force_ascii=False,     # 保留中文
    date_format="iso"      # date 用 ISO 格式字符串
)
```
